```python
print("PART A: DATA ACQUISITION")

print("Importing the required libraries")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from webencodings import Encoding

dataframe = pd.read_csv("E:/DEUS SDS 2024/second semester/machine leaning and techinues/Medical Cost Personal Dataset.csv")
print("\n")

print("Displaying the first 5 rows to understand the data structure")
display(dataframe.head())
print("\n")

print("Showing number of rows and columns in the dataset; observations and variables.")
display(dataframe.shape)
print("\n")

print("Displaying information about data types and non-null values")
display(dataframe.info())
print("\n")

print("Showing summary statistics for numerical variables")
display(dataframe.describe())
print("\n")
print(" in summary, the dataset represents the effects of age, gender, sex, bmi, children,  smoker  on the charges.")
print(" thus the variables that are independent are, sex, age , gender, bmii, children, smoker and the dependent is charges associated")       
print("\n")


print("PART B: DATA CLEANING")

print("Checking for missing values in each column")
display(dataframe.isnull().sum())
print(" there are no missing values.")
print("\n")
print("Checking for the number of duplicate records")
display(dataframe.duplicated().sum())


print("Removing the duplicate rows found in the dataset")
dataframe = dataframe.drop_duplicates()
print("\n")

print("Confirming new shape after removing duplicates")
display(dataframe.shape)
print("\n")

print("Outlier detection using the IQR method")
print("Defining function to detect outliers using the IQR method")
def detect_outliers(column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1 
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return dataframe[(dataframe[column] < lower) | (dataframe[column] > upper)]

print("Detecting outliers in BMI")
display(detect_outliers("bmi"))
print("\n")

print("Detecting outliers in Charges")
display(detect_outliers("charges"))
print("\n")

print(" Outliers were found in both the BMI and charges variables using the Interquartile Range (IQR) method.") 
print("For BMI, extreme values above 47 were noted, with the highest reaching 53.13. " )
print("For charges, high values above 35,000 were identified, with the highest exceeding 62,000")
print("Instead of removing the outliers, i kept them in the dataset the dataset represents real-world insurance records,")
print("and these extreme values are actual observations, not data entry mistakes.")
print("Excluding them would unduly lower variability and overlook high-risk groups that interest insurance companies.")
print("Also, from a business viewpoint, insurers need to consider clients with obesity or smoking-related health risks since they lead to much higher medical costs.") 
print("Keeping these outliers helps to ensure that the regression model accurately reflects the financial impact of such high-risk groups.")
print("In conclusion, the outliers were retained to keep the integrity of the dataset and to make sure the predictive")
print("model shows the full range of medical insurance charges, including extreme but realistic cases.")

print("\n")
    
print("Encoding Categorical Variables")
dataframe['sex'] = dataframe['sex'].map({'male': 1, 'female': 0})
dataframe['smoker'] = dataframe['smoker'].map({'yes': 1, 'no': 0})

print("Converting categorical variables into dummy variables")
dataframe_encoded = pd.get_dummies(dataframe, columns=['region'], drop_first=True)

print("Displaying first few rows after encoding")
display(dataframe.dtypes)
print("\n")

display(dataframe.head())
print("\n")

print(" in summary, encoding dummy variables mthd, ensures that categorical data is put into numerics, thus helping in perfect regression analysis.")
print(" also , it helps in droping the data that is categorical and helps in grouping the outcomes.")
print("\n")
print(" for scaling, it is not necessary because ordinary least squares regression is scale‑invariant.")
print(" in otherwords, when predictors are in categorical formart, the model can still fit.")
print("\n")

print("PART C: EXPLORATORY DATA ANALYSIS")

print("Plotting histogram to see distribution of medical charges")
plt.hist(dataframe['charges'], bins=30)
plt.title("Distribution of Charges")
plt.xlabel("Charges")
plt.ylabel("Frequency")
plt.show()
print("\n")
print(" to comment on the skewness, the histogram is rightly skewed. This indicates that most of the individuals pay a relatively lower charges.")
print("\n")

print("Creating scatter plot to examine relationship between age and charges")
plt.scatter(dataframe['age'], dataframe['charges'])
plt.title("Age vs Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
print("\n")

print("Creating scatter plot to examine relationship between BMI and charges")
plt.scatter(dataframe['bmi'], dataframe['charges'])
plt.title("BMI vs Charges")
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.show()
print("\n")

print("Boxplot to compare charges between smokers and non-smokers")
sns.boxplot(x=dataframe['smoker'], y=dataframe['charges'])
plt.title("Charges by Smoker Status")
plt.show()
print("\n")

print("Correlation matrix as heatmap")
sns.heatmap(dataframe_encoded.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
print("\n")

print(" According to the graphs obtained above, smokers have more charges. Thus this makes smoking strongly related to charges.")
print( "accorfing the graph of smokers vs charges, non-smokers indicated by 0 pay less money, and smokers indicated by 1, pay more money. Thus smokers pay more money than non-smokers")
print(" accordingly, age and bmi are approximmately lionear, though not perfectly linear.")
print("\n")


print("PART D: REGRESSION MODELING")
print("Importing regression libraries")
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import statsmodels.api as sm
import statsmodels.api as sm

print("Encoding the columns as numeric")
dataframe_encoded = dataframe_encoded.astype(float)
print("\n")

print("SIMPLE LINEAR REGRESSION")

X_simple = dataframe[['age']].astype(float)
y_simple = dataframe['charges'].astype(float)

X_simple = sm.add_constant(X_simple)

model_simple = sm.OLS(y_simple, X_simple).fit()
print("\n")

print("the summary of the model")
display(model_simple.summary())
print("\n")

print( " according to the model, the slope of the age and the charges is 257.2297. this implies that, for evry increase in age, medical charges are expected to rise by 257 units.")
print("\n")


print("MULTIPLE LINEAR REGRESSION")
print("\n")

print("Defining the predictors")
X = dataframe_encoded.drop('charges', axis=1)

print("Defining target")
y = dataframe_encoded['charges']

print("Converting everything to float")
X = X.astype(float)
y = y.astype(float)

print("Adding a constant")
X = sm.add_constant(X)

print("Fitting the model")
model = sm.OLS(y, X).fit()
print("\n")
print(" the summary of the model.")
display(model.summary())
print("\n")

print("Accordingly, variables(predictors) that have the p value less than 0.500 are considered significant, thus making the rest significant, while sex not significant") 
print("according to the analysis, smokers have  a great effect on charges compared to other predictors")
print(" smoking affects the charges, where smokers pay about 2.385e4( 23,580) units on average holding other facors constant. thus making it affect charges greatly compare with the rest predictors")
print("\n")


print("PART E: MODEL EVALUATION")

print("Splitting the data")
X_train, X_test, y_train, y_test = train_test_split(
    dataframe_encoded.drop('charges', axis=1).astype(float),
    dataframe_encoded['charges'].astype(float),
    test_size=0.2,
    random_state=42
)

print("Creating a model")
lr = LinearRegression()

print("Training the model")
lr.fit(X_train, y_train)

print("Making predictions")
y_pred = lr.predict(X_test)

print("Evaluation metrics")
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("\n")

print("RMSE:", rmse)
print("MAE:", mae)
print("R²:", r2)
print("\n")
print("\n")


print("Calculating the residuals using the (actual - predicted) method")
residuals = y_test - y_pred
print("\n")

print("Plotting residuals to check for homoscedasticity")
plt.scatter(y_pred, residuals)
plt.axhline(y=0)
plt.title("Residual Plot")
plt.xlabel("Predicted Charges")
plt.ylabel("Residuals")
plt.show()
print("\n")

print("Normality of Residuals")
print("Plotting histogram of residuals to check normality")
plt.hist(residuals, bins=30)
plt.title("Histogram of Residuals")
plt.show()
print("\n")

print("Checking Multicollinearity (VIF)")
print("Importing the VIF function")
from statsmodels.stats.outliers_influence import variance_inflation_factor
print("\n")

print("Adding a constant to predictors")
X_vif = sm.add_constant(dataframe_encoded.drop('charges', axis=1))

print("Creating a dataframe to store VIF values")
vif_data = pd.DataFrame()
vif_data["Feature"] = X_vif.columns

print("Calculating VIF for each feature")
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
print("\n")

print("Displaying VIF results")
display(vif_data)

print("Smoking status is the strongest factor.")
print(" from model, the regression coefficient for smokers is about 23,850, meaning that smokers pay much higher charges compared to non-smokers")  
print("\n")

print(" from the model, the coefficient for BMI is about 339, meaning that for every 1-unit increase in BMI,") 
print("medical insurance charges are expected to rise by approximately 339 units, holding other factors constant.")
print("\n")


print("Adjust Premiums for Smokers ")  
print("Since smoking increases charges by ~23,850, smokers should be charged higher premiums to reflect their higher risk. ") 
print("This ensures fairness and helps manage financial risk.")
print("\n")

print("Incentivize Healthy Lifestyles")  
print("Offer discounts or wellness programs for clients who maintain a healthy BMI")  
print("Since BMI increases charges by ~339 per unit, encouraging weight management can reduce costs.")
print("\n")

print("Age-Based Pricing Strategy.")  
print(" Age also contributes significantly with ab0ut 257 units per year thus")  
print("Premiums should gradually increase with age to account for higher expected medical costs, while still remaining fair.")
print("\n")
```
